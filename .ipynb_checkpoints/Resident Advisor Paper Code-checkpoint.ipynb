```python
## dataframe early cleaning ##
# import csv file 
df = pd.read_csv('resident-adv.csv')
# replace special characters in df, but keep hyphen for title column
df["style"].replace(regex=True,inplace=True,to_replace='[^A-Za-z0-9\s-]+',value=r'')
# replace strange label 'li\r\ndiv\r\nLabel ' with Self-released
df.replace(inplace=True, to_replace= 'li\r\ndiv\r\nLabel ', value='NA')
# capitalise column names
df.columns = ['Title', 'Artist', 'Single', 'Label', 'Record', 'Genre', 'Review Date', 'Single Release Date', 'RA Comments', 'Rating', 'Description', 'URL']
# change reviewed_date to datetime
df['Review Date'] = pd.to_datetime(df['Review Date'], format = '%Y-%m-%d')
df
```

```python
df["Genre"].value_counts()
```

```python
## grouping the releted genre labels ##

# group all techno house
df.replace(inplace=True, to_replace= ["Techno House", "House Techno", "Tech house"], value='Tech House')
# group all minimal 
df.replace(inplace=True, to_replace= ["House Minimal","Minimal House", "Minimal Techno", "Techno Minimal", "Minimal house", "Minimal techno", 'Minimal Tech House','Tech House Minimal'] , value='Minimal')
# group all Techno
df.replace(inplace=True, to_replace= ["Techno Ambient", "Techno Bass", "Techno Industrial", "Techno Dub Techno"] , value='Techno')
# group all deep house
df.replace(inplace=True, to_replace= ["House Deep House", "Deep House House", "Deep house"] , value='Deep House')
# group all experimental
df.replace(inplace=True, to_replace= ["Experimental", "Progressive", "Techno Experimental","Experimental Techno", "Experimental House", "House Experimental", "Experimental house", "Experimental techno", "Progressive House", "House Progressive", "Progressive house", "Progressive techno"] , value='Progressive')
# group all breakbeat
df.replace(inplace=True, to_replace= ["Breaks"] , value='Breakbeat')
# group all disco
df.replace(inplace=True, to_replace= ["House Disco", "Disco House"] , value='Disco')
# group all electro
df.replace(inplace=True, to_replace= ["Electronica", "Electro House", "Electro Techno", "House Electro", "Techno Electro", "House Electronica", "Techno Electronica" ] , value='Electro')
# group dubstep
df.replace(inplace=True, to_replace= ["Dub", "Dub Dubstep", "Grime Dubstep", "Dubstep Grime"] , value='Dubstep')
```

```python
### pivot tables ###
# group by genre for rating
genre_review_date_rating = df_top_10_genres.groupby(["Genre","Review Date"]).agg("mean").reset_index().sort_values(["Genre", "Review Date"])
# get year
genre_review_date_rating["Review Year"] = pd.DatetimeIndex(genre_review_date_rating['Review Date']).year
# get month
genre_review_date_rating["Review Month"] = pd.DatetimeIndex(genre_review_date_rating['Review Date']).month
# concatenate
genre_review_date_rating["Month/Year"] = genre_review_date_rating["Review Month"].astype(str) + '/' + genre_review_date_rating["Review Year"].astype(str)

# group by genre for count
genre_review_date_count = df_top_10_genres.groupby(["Genre","Review Date"]).agg("count").reset_index().sort_values(["Genre", "Review Date"]).iloc[:,0:3]
# get year
genre_review_date_count["Review Year"] = pd.DatetimeIndex(genre_review_date_rating['Review Date']).year
# get month
genre_review_date_count["Review Month"] = pd.DatetimeIndex(genre_review_date_rating['Review Date']).month
# concatenate
genre_review_date_count["Month/Year"] = genre_review_date_rating["Review Month"].astype(str) + '/' + genre_review_date_rating["Review Year"].astype(str)

## pivot by year ##
# with mean rating rounded to two dp
genre_year_rating_pivot = pd.pivot_table(genre_review_date_rating, index = "Review Year", columns = "Genre", values = "Rating", aggfunc = 'mean').round(2)
# with count
genre_year_count_pivot = pd.pivot_table(genre_review_date_count, index = "Review Year", values = "Title", columns = "Genre", aggfunc = 'sum').fillna(0).astype(int).round()
# combine the pivot tables 
combined_genre_year_table = '(' + genre_year_rating_pivot.astype(str) + ' , ' + genre_year_count_pivot.astype(str) + ')'

## same again but pivot by month/year ##
genre_rating_pivot_month = pd.pivot_table(genre_review_date_rating, index = ["Review Year", "Review Month"], columns = "Genre", values = "Rating", aggfunc = 'mean').round(2)
genre_count_pivot_month = pd.pivot_table(genre_review_date_count, index = ["Review Year", "Review Month"], values = "Title", columns = "Genre", aggfunc = 'count').fillna(0).astype(int).round()
combined_genre_table_month = '(' + genre_count_pivot_month.astype(str) + ' , ' + genre_rating_pivot_month.astype(str) + ')'

# reformat 2017 index label
as_list_2 = combined_genre_year_table.index.tolist()
idx_2 = as_list_2.index(2017)
as_list_2[idx_2] = '2017 (until October)'
combined_genre_year_table.index = as_list_2
```

```python
# Create dataframe of just House and Techno
df_house_techno = df_top_10_genres[df_top_10_genres["Genre"].isin(["House", "Techno"])]
```

```python
## Onto genre classification ## 

# set independent and dependent variables
X = df_house_techno["Description"]
y = df_house_techno["Genre"]

# convert series to numpy array 
X_array = X.values
y_array = y.values.reshape(-1,1)

# encode genre labels into labels that classifier can read
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_array_encoded = le.fit_transform(y_array)

# sklearn's word count vectoriser
from sklearn.feature_extraction.text import TfidfVectorizer

# instantiate with sk learn vectorise
# convert a collection of text documents to a matrix of token counts
vect = TfidfVectorizer(max_features = 5000)
# learn training data vocabulary, then use it to create a document-term matrix
X_dtm = vect.fit_transform(X_array).toarray()
```

```python
# number of features extracted
len(X_dtm[0])
```

```python
# split data into training and test set
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_dtm, y_array_encoded)
class_names = np.unique(y_array)
```

```python
from sklearn.grid_search import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_jobs=-1, max_features='sqrt') 

# Use a grid over parameters of interest
param_grid = { 
           "n_estimators" : [10, 50, 100, 200, 400, 800],
           "max_depth" : [1, 5, 10, 15, 20, 25, 30],
           "min_samples_leaf" : [1, 2, 4, 6]}

CV_rfc = GridSearchCV(estimator=rfc, param_grid=param_grid, cv= 10)
CV_rfc.fit(X_train, y_train)
print(CV_rfc.best_params_) 
```

```python
# fit model with optimal hyperparameters
rfc = RandomForestClassifier(n_jobs = -1, max_features = 'sqrt', n_estimators = 800, max_depth = 30, min_samples_leaf = 1)
rfc.fit(X_train, y_train)
```

```python
y_true, y_pred = y_test, rfc.predict(X_test)

from sklearn.metrics import classification_report
classification_report(y_true, y_pred, target_names=class_names)
```

```python
# plotting a confusion matrix
import itertools
from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(cm, classes,
                          normalize=True,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation = 45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

# Compute confusion matrix
cnf_matrix = confusion_matrix(y_test, y_pred)
np.set_printoptions(precision=2)

# Plot normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=class_names, normalize=True,
                      title='Normalized confusion matrix')

plt.show()
```
