```python
# import csv file after web scraping
df = pd.read_csv('resident-adv.csv')
```

```python
# replace special characters in df, keep hyphen as useful for title column
df.replace(regex=True,inplace=True,to_replace='[^A-Za-z0-9\s-]+',value=r'')
# capitalise column names
df.columns = ['Title', 'Artist', 'Single', 'Label', 'Record', 'Genre', 'Review Date', 'Single Release Date', 'RA Comments', 'Rating', 'Description', 'URL']
# group all techno house
df["Genre"].replace(inplace = True, to_replace= ['Tech House', 'House Tech House', "Techy House", "Techno House", "Tech house", "House Techno", "HouseTechno", "TechnoHouse"], value='House/Techno')
# group all minimal 
df["Genre"].replace(inplace = True, to_replace= ['Minimal Tech House','Tech House Minimal', "Techno Minimal", "House Minimal", "Minimal Techno", "Minimal techno", "Minimal House"] , value='Minimal')
# group all Techno
df["Genre"].replace(inplace = True, to_replace= ["Techno Electro", "Techno Ambient", "Techno Bass", "Techno Industrial", "Techno Dub Techno"] , value='Techno')
# group all deep house
df["Genre"].replace(inplace = True, to_replace= ["House Deep House"] , value='Deep House')
# group all house
df["Genre"].replace(inplace = True, to_replace= ["House Downtempo"] , value='House')
# group all experimental
df["Genre"].replace(inplace = True, to_replace= ["Techno Experimental", "House Experimental", "Progressive House", "Progressive house", "Experimental Techno", "Progressive", "Experimental", "Bass Techno", "Experimental Progressive", "Progressive house House", "Experimental club "] , value='Experimental/Progressive')
# group all breakbeat
df["Genre"].replace(inplace = True, to_replace= ["Breaks"] , value='Breakbeat')
# group all disco
df["Genre"].replace(inplace = True, to_replace= ["House Disco", "Disco House"] , value='Disco')
# group all electro
df["Genre"].replace(inplace = True,to_replace= ["Electronica", "House Electro", "Electro Techno", "House Electronica", "Electro House"] , value='Electro')
# group bass
df["Genre"].replace(inplace = True, to_replace= ["UK Bass", ] , value='Bass')
```

```python
## pivot tables of rating, year and genre ##
genre_review_date_rating = df_top_7_genres.groupby(["Genre","Review Date"]).agg("mean").reset_index().sort_values(["Genre", "Review Date"])
genre_review_date_rating["Review Year"] = pd.DatetimeIndex(genre_review_date_rating['Review Date']).year
# with mean rating rounded to two dp
genre_year_rating_pivot = pd.pivot_table(genre_review_date_rating, index = "Review Year", columns = "Genre", values = "Rating", aggfunc = 'mean').round(2)
# with count
genre_year_rating_pivot_count = pd.pivot_table(genre_review_date_rating, index = "Review Year", values = "Rating", columns = "Genre", aggfunc = 'count').fillna(0).astype(int).round()
# combine the pivot tables 
combined_genre_year_table = '(' + genre_year_rating_pivot.astype(str) + ' , ' + genre_year_rating_pivot_count.astype(str) + ')'
```

```python
## add month with year to genre
genre_review_date_rating["Review Month"] = pd.DatetimeIndex(genre_review_date_rating['Review Date']).month


genre_year_rating_pivot_month = pd.pivot_table(genre_review_date_rating, index = ["Review Year", "Review Month"], columns = "Genre", values = "Rating", aggfunc = 'mean').round(2)
genre_year_rating_pivot_count_month = pd.pivot_table(genre_review_date_rating, index = ["Review Year", "Review Month"], values = "Rating", columns = "Genre", aggfunc = 'count').fillna(0).astype(int).round()
```

```python
## ONTO THE DESCRIPTION ANALYSIS ##
# create smaller df with just description and genre label
top_7_genres_simple = df_top_7_genres[["Description", "Genre"]]

# set independent and dependent variables
X = top_7_genres_simple["Description"]
y = top_7_genres_simple["Genre"]

# convert series to numpy array 
X_array = X.values
y_array = y.values.reshape(-1,1)
```

```python
# convert genres in labels that tpot can read
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_array_encoded = le.fit_transform(y_array)
```

```python
# sklearn's word count vectoriser
from sklearn.feature_extraction.text import TfidfVectorizer

# instantiate with sk learn vectorise
# convert a collection of text documents to a matrix of token counts
vect = TfidfVectorizer(max_features = 1000)
# learn training data vocabulary, then use it to create a document-term matrix
X_dtm = vect.fit_transform(X_array).toarray()
```

```python
# split data into training and test set
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_dtm, y_array_encoded)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```
